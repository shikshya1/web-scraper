```python
def extract_rating(rating_div,x):
    review=""
    rating="0"
    try:
        Reviewer = rating_div.find_all("span", {"class": "ui_bubble_rating bubble_50"})[x] 
        rating = "5"
    except:
        pass
        
    try:
        Reviewer = rating_div.find_all("span", {"class": "ui_bubble_rating bubble_40"})[x] 
        rating = "4"
    except:
        pass
    
    try:
        Reviewer = rating_div.find_all("span", {"class": "ui_bubble_rating bubble_30"})[x] 
        rating = "3"
    except:
        pass
        
    try:
        Reviewer = rating_div.find_all("span", {"class": "ui_bubble_rating bubble_20"})[x] 
        rating = "2"
    except:
        pass
        
    try:
        Reviewer = rating_div.find_all("span", {"class": "ui_bubble_rating bubble_10"})[x] 
        rating = "1"
    except:
        pass

    

    return rating
```
